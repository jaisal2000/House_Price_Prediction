```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/House Price India.csv"
```

```python
data = pd.read_csv(path)
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
stats = data["Price"].describe().reset_index()
stats["Price"]=round(stats["Price"],2)
```

```python
stats
```

```python
data.isna().sum()
```

```python
# for na values
data.dropna(inplace = True)
# for duplicate values
data.drop_duplicates(inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending = False).plot(kind = "bar")
plt.title("condition of house v/s Price")
plt.ylabel("Mean Price")
plt.show()
```

```python
data.columns
```

```python
X = data[["number of bedrooms","number of bathrooms",'living area','condition of the house','Number of schools nearby', 'Distance from the airport']]
Y = data[["Price"]]
```

```python
X
```

```python
Y
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid={
    "criterion":["mse","friedman_mse","mae"],
    "splitter" : ["best","random"],
    "max_depth" : [None, 10, 20, 30, 40, 50],
    "min_samples_split" : [2,5,10],
"min_samples_leaf" : [1,2,4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree= GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(X_train,Y_train)
```

```python
grid_tree.best_params_
```

```python
 tree_preds= grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(Y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {
    "max_depth" : [5,10,15],
"n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(X_train, Y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(X_test)
```

```python
mean_absolute_error(Y_test,rfrpredictions)
```
